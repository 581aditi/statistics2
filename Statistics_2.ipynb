1. What is hypothesis testing in statistics?
-> In hypothesis testing, we start with a null hypothesis (H₀), which represents a statement of no effect or no difference, and an alternative hypothesis (H₁ or Ha), which represents a statement of effect or difference. The goal is to use sample data to determine whether there is enough statistical evidence to reject the null hypothesis in favor of the alternative hypothesis.

2. What is null hypothesis, and how does it differ from alternative hypothesis
-> In statistical hypothesis testing, the null hypothesis (H₀) and the alternative hypothesis (H₁ or Ha) are two mutually exclusive statements that researchers use to test assumptions about a population parameter.

3. What is the significance level in hypothesis testing, and why is it important?
-> In hypothesis testing, the significance level, denoted as α (alpha), is a threshold set by the researcher before conducting the test. It represents the probability of rejecting the null hypothesis when it is actually true, known as a Type I error. In other words, it's the risk you're willing to take for making a false positive conclusion.

Importance of the Significance Level:
Controlling Type I Errors: By setting a significance level, researchers control the likelihood of making incorrect conclusions.
Balancing Errors: Choosing the right significance level helps balance the risk of Type I errors against the need for sensitivity to detect actual effects.
Contextual Relevance: The choice of significance level depends on the context of the study and the consequences of making errors.
In medical research, a lower significance level (e.g., 0.01) is often used to minimize the risk of false positives.
In exploratory research, a higher significance level (e.g., 0.10) might be acceptable.

4. What does a P-value represent in hypothesis testing?
-> The p-value indicates how likely it is to observe the collected data, or something more extreme, under the assumption that the null hypothesis is correct. A smaller p-value suggests stronger evidence against the null hypothesis. For example, a p-value of 0.01 means there's a 1% chance of observing the data if the null hypothesis is true.

5. How do you interpret the P-value in hypothesis testing?
-> In hypothesis testing, the p-value is a statistical measure that helps determine the strength of the evidence against the null hypothesis (H₀). It represents the probability of obtaining results at least as extreme as those observed, assuming the null hypothesis is true.
Interpreting the P-Value:
P-value ≤ α (e.g., 0.05): Reject the null hypothesis; the result is statistically significant.
P-value > α: Fail to reject the null hypothesis; the result is not statistically significant.
Here, α (alpha) is the pre-determined significance level, commonly set at 0.05, representing a 5% risk of concluding that a difference exists when there is none.

6. What are Type1 and Type2 errors in hypothesis testing?
-> Type I Error (False Positive):
Definition: Occurs when the null hypothesis (H₀) is true, but is incorrectly rejected.
Consequence: Concluding that there is an effect or difference when, in reality, there isn't.
Probability: Denoted by α (alpha), often set at 0.05, indicating a 5% risk of making this error.
Example: A clinical trial concludes that a new drug is effective when it actually isn't.
Type II Error (False Negative):
Definition: Occurs when the null hypothesis is false, but is incorrectly not rejected.
Consequence: Failing to detect an effect or difference that actually exists.
Probability: Denoted by β (beta). The power of a test is calculated as 1 - β, representing the probability of correctly rejecting a false null hypothesis.
Example: A clinical trial fails to detect the effectiveness of a new drug that actually works.

7. What is a difference between a one-tailed and two-tailed testing in hypothesis testing?
-> One-Tailed Test:
Definition: Tests for the possibility of an effect in one direction (either greater than or less than a certain value).
Investopedia
Critical Region: Located entirely in one tail of the probability distribution
Significance Level (α): All of α is allocated to one tail. For example, with α = 0.05, the critical region is in the top 5% or bottom 5% of the distribution, depending on the direction
Use Case: Appropriate when the research hypothesis predicts a specific direction of the effect.
Example: Testing if a new drug increases recovery time compared to an existing drug.
Two-Tailed Test:
Definition: Tests for the possibility of an effect in both directions (either greater than or less than a certain value).
Critical Regions: Located in both tails of the probability distribution
Significance Level (α): α is split between the two tails. For example, with α= 0.05, each tail has a 2.5% critical region, totaling 5%.
Use Case: Appropriate when the research hypothesis does not predict the direction of the effect.
Example: Testing if a new drug has a different effect on recovery time compared to an existing drug, without specifying whether it's better or worse.

8. What is the Z-test,and when it is used in hypothesis testing?
-> A Z-test is a statistical method used to determine whether there is a significant difference between the sample mean and the population mean, or between the means of two samples, under certain conditions.
A Z-test is applicable when:
Sample Size: The sample size is large (typically n>30)
Population Standard Deviation: The population standard deviation (σ) is known
Data Distribution: The data is approximately normally distributed, or the sample size is large enough for the Central Limit Theorem to apply
Under these conditions, the Z-test uses the standard normal distribution to calculate the Z-score, which measures how many standard deviations the sample mean is away from the population mean

9. How do you calculate the Z-score,and what does it represent in hypothesis testing?
-> In hypothesis testing, the Z-score (also known as the standard score) quantifies how many standard deviations a data point is from the population mean. It standardizes scores on a common scale, facilitating comparisons across different datasets.
The Z-score is calculated using the formula:
Z= X−μ/σ
​Where:
X = Raw score (individual data point)
μ = Population mean
σ = Population standard deviation

10. What is the T-distribution, and when it should be used instead of the normal distribution?
-> The t-distribution is a probability distribution that is symmetric and bell-shaped, similar to the normal distribution but with heavier tails. It is used in statistical analyses, particularly when dealing with small sample sizes or when the population standard deviation is unknown
The t-distribution should be used instead of the normal distribution in the following situations:
Small Sample Size: When the sample size is small (typically less than 30), the t-distribution provides a more accurate estimate of the population parameters.
Unknown Population Standard Deviation: If the population standard deviation is unknown and is estimated from the sample data, the t-distribution is appropriate.
JMP Statistical Discovery
Approximately Normal Population: The population from which the sample is drawn should be approximately normal. If the sample size is small and the population distribution is distinctly not normal (e.g., heavily skewed or contains outliers), using the t-distribution can lead to unreliable results. In such cases, non-parametric tests may be more appropriate or transformations of the data may be required.

14. What is a confidence interval,and how is it used to interpret statistical results?
-> A confidence interval (CI) is a statistical tool that provides a range of values within which we can be reasonably certain that a population parameter (such as a mean or proportion) lies. Rather than offering a single point estimate, a confidence interval conveys the uncertainty inherent in sampling.
Interpreting Confidence Intervals:
Precision of Estimates: A narrower CI indicates a more precise estimate of the population parameter, while a wider CI suggests greater uncertainty .
QuickTakes
Statistical Significance: If a CI for a mean difference includes zero, it implies that there is no statistically significant difference between the groups at the chosen confidence level.
Hypothesis Testing: Confidence intervals complement p-values in hypothesis testing. For instance, if a 95% CI does not include the null hypothesis value (e.g., zero), it suggests that the null hypothesis can be rejected at the 5% significance level.

15. What is the margin of error,and how does it affect the confidence interval?
-> The margin of error (MoE) quantifies the uncertainty in a sample estimate, indicating how much the sample statistic (e.g., sample mean or proportion) might differ from the true population parameter. It is a crucial component in constructing confidence intervals, which provide a range of plausible values for the population parameter.
Factors affecting the margin of error include:
Sample Size (n): Increasing the sample size reduces the margin of error, enhancing the precision of the estimate.
Population Variability (σ): Higher variability increases the margin of error, indicating less certainty about the estimate.
Confidence Level: A higher confidence level (e.g., 99% vs. 95%) requires a larger margin of error to ensure greater certainty that the interval contains the true parameter.

17. What is the Chi-square distribution,and when is it used?

Variance Estimation: In cases where the underlying distribution is normal, the Chi-square distribution is used to construct confidence intervals for the population variance.

20. What is an ANOVA test,and what are its assumptions?
-> ANOVA assesses the impact of one or more factors by comparing the means of different samples. It evaluates the null hypothesis that all group means are equal against the alternative hypothesis that at least one group mean is different. The test calculates an F-statistic, which is the ratio of the variance between the group means to the variance within the groups. A higher F-statistic suggests a greater likelihood that at least one group mean is different.
Assumptions of ANOVA:
Normality: The data in each group should be approximately normally distributed. This assumption ensures that the F-statistic follows the correct distribution.
Independence: Observations within each group must be independent of each other. This means that the data points are not related or influenced by each other.

Homogeneity of Variances (Homoscedasticity): The variance within each group should be approximately equal. Unequal variances can lead to biased results.

No Significant Outliers: Outliers can distort the results of ANOVA. It's important to detect and handle outliers appropriately before conducting the test.

21. What are the different types of ANOVA test?
-> Different types of ANOVA test are:
1. One-Way ANOVA:
Purpose: Compares the means of three or more independent groups based on one factor.
Example: Assessing the impact of three different fertilizers on plant growth.
2. Two-Way ANOVA:
Purpose: Examines the influence of two independent variables on one dependent variable, and also assesses the interaction between these variables.
3. Repeated Measures ANOVA:
Purpose: Compares means across multiple time points or conditions within the same subjects.
Example: Evaluating blood pressure readings before, during, and after treatment in the same group of patients.
4. Factorial ANOVA:
Purpose: Assesses the effect of two or more factors simultaneously and their interactions on a dependent variable.
Example: Investigating the combined effects of diet (vegetarian vs. non-vegetarian) and exercise (light vs. intense) on weight loss.
5. Mixed-Design ANOVA:
Purpose: Combines elements of both repeated measures and factorial ANOVA, analyzing both within-subjects and between-subjects factors.
Example: Studying the impact of a new teaching method (between-subjects factor) and time (pre-test vs. post-test) on student performance.
Assumptions: Normality, independence, and homogeneity of variances.
6. Multivariate Analysis of Variance (MANOVA):
Purpose: Extends ANOVA by analyzing multiple dependent variables simultaneously.
Example: Assessing the effects of different diets on both weight and cholesterol levels.
7. ANOVA on Ranks:
Purpose: A non-parametric alternative to ANOVA, used when data do not meet the assumptions of normality.
Example: Comparing customer satisfaction ratings across different stores when data are ordinal.
8. Scheirer–Ray–Hare Test:
Purpose: A non-parametric test used to analyze ranked data from factorial designs, serving as an alternative to factorial ANOVA when data are not normally distributed.
Example: Examining the effects of two different fertilizers and two irrigation methods on plant growth using ranked data.

22. What is the F-test,and how does it relate to hypothesis testing?
-> The F-test is a statistical method used to compare variances between two or more groups. It evaluates whether the observed variances are significantly different, which can indicate differences in population means. The test computes an F-statistic, the ratio of variances, and compares it to a critical value from the F-distribution to determine statistical significance.

In hypothesis testing, the F-test is employed to assess the null hypothesis (H₀) that two populations have equal variances. The alternative hypothesis (H₁) posits that the variances are unequal. A significant F-statistic leads to rejecting H₀, suggesting that the populations have different variances. This test is particularly useful in comparing models or assessing the equality of variances before performing other analyses like t-tests.

```python
#2 Generate a discrete uniform distribution using python and plot the probability mass function(PMF)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for the uniform distribution
low, high = 1, 6

# Create a discrete uniform distribution
distribution = randint(low, high + 1)

# Generate the range of outcomes
x = np.arange(low, high + 1)

# Compute the PMF
pmf_values = distribution.pmf(x)

# Plot the PMF
plt.stem(x, pmf_values, basefmt=", " use_line_collection=True)
plt.title(f'Discrete Uniform Distribution PMF: U({low}, {high})')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
#3 Write a python function to calculate the probability distribution function (PDF) of bernoulli distribution
def bernoulli_pmf(k, p):
    """
    Calculate the Probability Mass Function (PMF) of a Bernoulli distribution.

    Parameters:
    - k: Outcome (0 or 1)
    - p: Probability of success (0 <= p <= 1)

    Returns:
    - PMF value for the given outcome k
    """
    if k not in [0, 1]:
        raise ValueError("k must be either 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1 inclusive")

    return p**k * (1 - p)**(1 - k)

# Example usage:
p = 0.7
for k in [0, 1]:
    print(f"P(X = {k}) = {bernoulli_pmf(k, p):.4f}")
```

```python
#9 Write a python function to calculate Z-score from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate the Z-scores for a given dataset and plot the histogram of the Z-scores.

    Parameters:
    - data: List or NumPy array of numerical values.
    """
    # Convert data to a NumPy array if it's a list
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-Scores')
    plt.xlabel('Z-Score')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage
data = [70, 80, 90, 100, 110]
z_scores = calculate_and_plot_z_scores(data)
print("Z-scores:", z_scores)
```

```python
#13 Generate random variables and calculate their corresponding probabilities using the binomial distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def generate_binomial_data(n, p, size):
    """
    Generate random variables from a binomial distribution and calculate their probabilities.

    Parameters:
    - n: number of trials
    - p: probability of success on each trial
    - size: number of random variables to generate

    Returns:
    - random_variables: generated random variables
    - probabilities: corresponding probabilities
    """
    # Generate random variables
    random_variables = np.random.binomial(n, p, size)

    # Calculate probabilities using the binomial PMF
    probabilities = binom.pmf(random_variables, n, p)

    return random_variables, probabilities

# Parameters
n = 10  # number of trials
p = 0.5  # probability of success
size = 1000  # number of random variables to generate

# Generate data
random_vars, probs = generate_binomial_data(n, p, size)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(random_vars, probs, alpha=0.6, color='blue')
plt.title(f'Binomial Distribution: n={n}, p={p}')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
#14 Write a python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, x):
    """
    Calculate the Z-score for a given data point x.

    Parameters:
    - data: List or NumPy array of numerical values.
    - x: The data point for which to calculate the Z-score.

    Returns:
    - z_score: The calculated Z-score.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_score = (x - mean) / std_dev
    return z_score

def plot_standard_normal_distribution():
    """
    Plot the standard normal distribution (mean=0, std=1).
    """
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')
    plt.title('Standard Normal Distribution (μ=0, σ=1)')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage
data = [70, 75, 80, 85, 90, 95, 100, 105, 110, 115]
x = 90  # Data point to calculate Z-score

z_score = calculate_z_score(data, x)
print(f"Z-score for {x} is: {z_score:.2f}")

# Plotting the standard normal distribution
plot_standard_normal_distribution()
```

```python
#5 Create a python function that calculates and visualizes type 1 and type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def simulate_hypothesis_test(pop_mean, alt_mean, std_dev, sample_size, alpha=0.05, num_simulations=10000):
    # Initialize counters for errors
    type_1_errors = 0
    type_2_errors = 0

    # Perform simulations
    for _ in range(num_simulations):
        # Sample data under H0 (null hypothesis)
        sample_h0 = np.random.normal(loc=pop_mean, scale=std_dev, size=sample_size)
        sample_mean_h0 = np.mean(sample_h0)
        z_stat_h0 = (sample_mean_h0 - pop_mean) / (std_dev / np.sqrt(sample_size))
        p_value_h0 = 2 * (1 - norm.cdf(abs(z_stat_h0)))

        # Sample data under H1 (alternative hypothesis)
        sample_h1 = np.random.normal(loc=alt_mean, scale=std_dev, size=sample_size)
        sample_mean_h1 = np.mean(sample_h1)
        z_stat_h1 = (sample_mean_h1 - pop_mean) / (std_dev / np.sqrt(sample_size))
        p_value_h1 = 2 * (1 - norm.cdf(abs(z_stat_h1)))

        # Check for Type I error (rejecting H0 when H0 is true)
        if p_value_h0 < alpha:
            type_1_errors += 1

        # Check for Type II error (failing to reject H0 when H1 is true)
        if p_value_h1 >= alpha:
            type_2_errors += 1

    # Calculate error rates
    type_1_error_rate = type_1_errors / num_simulations
    type_2_error_rate = type_2_errors / num_simulations

    # Output results
    print(f"Type I Error Rate: {type_1_error_rate:.4f}")
    print(f"Type II Error Rate: {type_2_error_rate:.4f}")

    # Visualization
    fig, ax = plt.subplots(figsize=(10))
```

```python
#9 Write a python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
from scipy import stats

def calculate_confidence_interval(sample_data, confidence_level=0.95):
    # Sample statistics
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation
    sample_size = len(sample_data)

    # Degrees of freedom
    df = sample_size - 1

    # t-value for the given confidence level
    t_value = stats.t.ppf((1 + confidence_level) / 2, df)

    # Standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Margin of error
    margin_of_error = t_value * standard_error

    # Confidence interval
    lower_limit = sample_mean - margin_of_error
    upper_limit = sample_mean + margin_of_error

    return lower_limit, upper_limit

# Example usage
sample_data = [45, 52, 48, 55, 49, 51, 53, 47, 50, 54]
ci = calculate_confidence_interval(sample_data)
print(f"95% Confidence Interval: {ci}")

#Significance of Confidence Intervals
A confidence interval offers several key insights:
Uncertainty Quantification: It provides a range that likely contains the true population parameter, reflecting the uncertainty inherent in sample-based estimates.
Decision Making: In fields like medicine and economics, confidence intervals help in making informed decisions by understanding the precision of estimates.
Statistical Significance: If a confidence interval for a difference between two groups does not include zero, it suggests a statistically significant difference between them.
Misconceptions: It's crucial to note that a 95% confidence interval does not mean there's a 95% probability that the true parameter lies within the interval. Instead, it means that if we were to take 100 different samples and compute a CI for each,
approximately 95 of them would contain the true parameter .
```

```python
#10 Write a python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy import stats

def calculate_margin_of_error(sample_data, confidence_level=0.95):
    # Sample statistics
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation
    sample_size = len(sample_data)

    # Degrees of freedom
    df = sample_size - 1

    # t-value for the given confidence level
    t_value = stats.t.ppf((1 + confidence_level) / 2, df)

    # Standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Margin of error
    margin_of_error = t_value * standard_error

    return margin_of_error

# Example usage
sample_data = [45, 52, 48, 55, 49, 51, 53, 47, 50, 54]
margin_of_error = calculate_margin_of_error(sample_data)
print(f"Margin of Error: {margin_of_error:.2f}")
```

```python
#11 Implement a Bayesian inference method using Bayes theorem in python and explain the process
def bayes_theorem(prior, likelihood, false_positive_rate, evidence):
    # Calculate the marginal likelihood (P(B))
    marginal_likelihood = (likelihood * prior) + (false_positive_rate * (1 - prior))

    # Apply Bayes' Theorem
    posterior = (likelihood * prior) / marginal_likelihood
    return posterior

# Given probabilities
prior = 0.01  # P(Disease)
likelihood = 0.95  # P(Pos|Disease)
false_positive_rate = 0.05  # P(Pos|No Disease)

# Calculate posterior probability
posterior = bayes_theorem(prior, likelihood, false_positive_rate, evidence=None)
print(f"The probability of having the disease given a positive test result is: {posterior:.4f}")
```
